```python
import PIL.Image as image
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
gray_img = image.open("images/spine.tif")

plt.imshow(gray_img, cmap="gray")
plt.show()
```

```python
k = 50
img = np.array(gray_img)

H,W = img.shape
print("Height: ", H)
print("Width: ", W)

img_invert = np.zeros((H,W))

for i in range(H):
    for j in range(W):
        img_invert[i,j] = 255 - img[i, j]
           

# binary_img = np.where(img>k,255,0)

plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(img, cmap="gray")
plt.subplot(1,2,2)
plt.title("Invert Imgae")
plt.imshow(img_invert, cmap="gray")
plt.show()
```

```python
img_float = np.array(gray_img).astype('float32')
H,W = img_float.shape
output = np.zeros((H,W))


for i in range(H):
    for j in range(W):
        img_float[i,j] = img_float[i,j] / 255
        output[i,j] = 1 * math.log(1+img_float[i,j])
        output[i,j] = output[i,j] * 255

output.astype('uint8')
plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(gray_img, cmap="gray")
plt.subplot(1,2,2)
plt.title("Logarithm Imgae")
plt.imshow(output, cmap="gray")
plt.show()
```
