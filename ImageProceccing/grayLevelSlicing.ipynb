```python
import PIL.Image as image
import matplotlib.pyplot as plt
import numpy as np
```

```python
gray_img = image.open("./images/dollar.tif")

plt.imshow(gray_img, cmap="gray")
plt.show()
```

```python
img = np.array(gray_img)

H,W = img.shape
print("Height: ", H)
print("Width: ", W)

img_slice = np.zeros((H,W))
A=100
B=180
for i in range(H):
    for j in range(W):
        if img[i, j]<=B and img[i, j]>=A:
            img_slice[i,j] = 255
        else:
            img_slice[i,j] = img[i, j]
           

plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(img, cmap="gray")
plt.subplot(1,2,2)
plt.title("sliced image")
plt.imshow(img_slice, cmap="gray")
plt.show()
```

```python
img = np.array(gray_img)

H,W = img.shape
print("Height: ", H)
print("Width: ", W)

img_sliced_bg = np.zeros((H,W))
A=100
B=180
for i in range(H):
    for j in range(W):
        if B>=img[i, j] and img[i, j]>=A:
            img_sliced_bg[i,j] = 220
        else:
            img_sliced_bg[i,j] = 0
           

plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(img, cmap="gray")
plt.subplot(1,2,2)
plt.title("sliced image bg supp Imgae")
plt.imshow(img_sliced_bg, cmap="gray")
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
plt.title("Original Image")
plt.imshow(img, cmap="gray")
plt.subplot(2,2,2)
plt.title("sliced Imgae")
plt.imshow(img_slice, cmap="gray")
plt.subplot(2,2,3)
plt.title("sliced image(background suppressed)")
plt.imshow(img_sliced_bg, cmap="gray")
plt.show()
```
