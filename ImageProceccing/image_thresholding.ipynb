```python
import PIL.Image as image
import matplotlib.pyplot as plt
import numpy as np

```

```python
# Read a grayscale image
gray_img = image.open("images/hand_xray.jpg")

plt.imshow(gray_img, cmap="gray")
plt.show()
```

```python
# Set a threshold value, k = 128
k = 128
img = np.array(gray_img)

H,W = img.shape
print("Height: ", H)
print("Width: ", W)


# Apply binary thresholding using:
# if r > k  s = 255
# else  s = 0

binary_img = np.where(img>k,255,0)

# Display the binary image
plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(img, cmap="gray")
plt.subplot(1,2,2)
plt.title("Threshold Imgae")
plt.imshow(binary_img, cmap="gray")
plt.show()
```

```python
# Create three binary images using different thresholds: k = 50, 100, 150
sizes = [50,100,150]
binary_images = []
for i in sizes:
    binary_images.append(np.where(img>i,255,0))
```

```python
# Compare the results
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
plt.title("Original")
plt.imshow(img, cmap="gray")
for i in range(3):
    plt.subplot(2,2,i+2)
    plt.title(f"k={sizes[i]}")
    plt.imshow(binary_images[i], cmap="gray")
plt.show()
```
