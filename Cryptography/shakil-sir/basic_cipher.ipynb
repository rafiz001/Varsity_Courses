```python
def encrypt(plain_text):
    cipher_text = ""
    for ch in plain_text:
        if ch.isalpha():
            ch = chr((ord(ch)+3))
            if ch.isalpha()!= True:
                ch = chr(ord(ch)-26)
        cipher_text += ch
    return cipher_text
```

```python
def decrypt(cipher_text):
    plain_text = ""
    for ch in cipher_text:
        if ch.isalpha():
            ch = chr((ord(ch)-3))
            if ch.isalpha()!= True:
                ch = chr(ord(ch)+26)
        plain_text += ch
    return plain_text
```

```python
plain = "I am Rafiz"
print(f"Plain Text: {plain}")
encrypted = encrypt(plain)
print(f"Encrypted Text: {encrypted}")
print(f"Decrypted Text: {decrypt(encrypted)}")
```
