```python
import numpy as np
import math
```

```python
def encrypt(plaintext, width = 4):
    cipher = ""
    length = len(plaintext)
    col = width
    row = math.ceil(length/width)
    grid = np.empty([row,col],"str_")
    k=0
    for i in range(row):
        for j in range(col):
            if(k<length):
                grid[i,j] = plaintext[k]
                k=k+1
    for j in range(col):
        for i in range(row):
            cipher+=grid[i,j]
    return cipher
```

```python
def decrypt(encrypted, width = 4):
    plain = ""
    length = len(encrypted)
    col = width
    row = math.ceil(length/width)
    grid = np.empty([row,col],"str_")
    empty = (row*col)-length
    k=0
    for j in range(col):
        for i in range(row):
            if not (i==(row-1) and j>=(col-empty) ):
                if(k<length):
                    grid[i,j] = encrypted[k]
                    k=k+1
    for i in range(row):
        for j in range(col):
            plain+=grid[i,j]
    return plain
```

```python
plain = "DEPARTMENT"
encrypted = encrypt(plain)
decrypted = decrypt(encrypted)
print(f"Plain=     {plain} and length = {len(plain)}")
print(f"Encrypted= {encrypted} and length = {len(encrypted)}")
print(f"Decrypted= {decrypted} and length = {len(decrypted)} ")
```
