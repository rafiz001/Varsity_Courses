```python
import numpy as np
import math
```

```python
def encrypt(plaintext, width = 4):
    length = len(plaintext)
    cipher = ""
    
    for i in range(width):
        for j in range(i,length,width):
            cipher+=plaintext[j]
    return cipher
```

```python
def decrypt(encrypted, width = 4):
    plain = ""
    length = len(encrypted)
    col = width
    row = math.ceil(length/4)
    grid = np.empty([row,col],"str_")
    empty = (row*col)-length
    k=0
    for j in range(col):
        for i in range(row):
            if not (i==(row-1) and j>=(col-empty) ):
                if(k<length):
                    grid[i,j] = encrypted[k]
                    k=k+1
    for i in range(row):
        for j in range(col):
            plain+=grid[i,j]
    return plain
```

```python
plain = "DEPARTMENT"
encrypted = encrypt(plain)
print(f"Encrypted= {encrypted}")
print(f"Decrypted= {decrypt(encrypted)}")
```
